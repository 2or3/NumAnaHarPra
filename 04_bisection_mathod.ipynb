方程式 $e^{-x^2} − \sin{x} = 0$ の解を下記の手順でそれぞれ二分法で求めよう。

(1) $f(x) = e^{-x^2} − \sin{x}$ として、 $y = f(x)$ のグラフを $ 0 \leq x \leq 10 $ の範囲で描け.

```julia
using PyPlot
```

```julia
x = range(0,stop=10,length=1000);
y = exp.(map(xx -> -xx^2, x)) - sin.(x);
plot(x, y, color="b", linewidth=2.0)
```
