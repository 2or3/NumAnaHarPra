方程式 $e^{-x^2} − \sin{x} = 0$ の解を下記の手順でそれぞれニュートン
法で求めよう。

(i) $f(x) = e^{-x^2} − \sin{x}$ として、 $y = f(x)$ のグラフを $ 0 \leq x \leq 10 $ の範囲で描け.

```julia
using PyPlot
```

```julia
function f(x)
    return exp(-x^2) - sin(x)
end
x=-1.:.001:10.; plot(x, @. f(x))
```

```julia
function f_1(x)
    return -2*x*exp(-x^2) - cos(x)
end
x=-1.:.001:10.; plot(x, @. f(x))
```

(ii) この方程式の $x > 0$ の解のうち、小さい方から2番目と3番目の2つの解をニュートン法で求めよ。  
それぞれの方法で使う初期値はグラフから判断して選ぶ。  
収束判定の $\varepsilon$ は $\varepsilon = 1.0 × 10^{−15}$ (c のプログラムでは $1.0e − 15$ と書けばよい）である。

```julia
function newtown(x)
    e = 1.0e-15
    x1 = x - f(x) / f_1(x)
    sol_list = []
    while abs(x1-x) > e
        x = x1
        x1 = x - f(x) / f_1(x)
        push!(sol_list, x)
    end
    return sol_list, x
end
```

```julia
sol_list3, c3 = newtown(3)
sol_list4, c4 = newtown(4)
sol_list5, c5 = newtown(5)
sol_list6, c6 = newtown(6)
sol_list7, c7 = newtown(7)
```
